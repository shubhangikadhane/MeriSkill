#### Outlier Removal 

```python
avg = df["BloodPressure"].mean()
df["BloodPressure"] = df["BloodPressure"].apply(lambda v:avg if v==0 else v)
```

```python
plt.figure(figsize = (8,5))
sns.distplot(df["BloodPressure"] , bins = 30,color = "Orange")
plt.title(" Distribution of BloodPressure")
```

```python
bmi_avg = df["BMI"].mean()
df["BMI"] = df["BMI"].apply(lambda v:avg if v==0 else v)
```

```python
plt.figure(figsize = (8,5))
sns.displot(df, x="BMI",color = "Purple",kind= "kde")
plt.title(" Distribution of BMI")
```

* As seen earlier there are amount whivh varies largely from rest of amount .It may not me outlier in real case bot for better modelling we need  to remove it.
* In above varible i replaced outlier value with mean. Noe lets use z-score technique to see outlier in other feature variable.

```python
df["z_score_bp"]=(df.BloodPressure-df.BloodPressure.mean()/df.BloodPressure.std())
df["z_score_Glucose"]=(df.Glucose-df.Glucose.mean()/df.Glucose.std())
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop(columns=["z_score_bp",	"z_score_Glucose"],inplace=True)
df.head()
```

### Split Data


```python
X= df.drop(columns="Outcome")
Y = df["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=30)
```

## Model building


### Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
M_l = LogisticRegression()
```

```python
M_l.fit(x_train , y_train)
```

```python
y_predicted_l = M_l.predict(x_test)
```

```python
M_l.score(x_test , y_test)
```

###### Evaluate the model performance using matrics like Accuracy,Confusion Matrix and Classification Report

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
Accuracy = accuracy_score(y_test,y_predicted_l)
print('Accuracy:',Accuracy)
```

```python
Confusion_matrix = confusion_matrix(y_test,y_predicted_l)
print('Confusion_matrix: \n',Confusion_matrix)
```

```python
Classification_report = classification_report(y_test,y_predicted_l)
print('Classification_report: \n',Classification_report)
```

```python
from sklearn.metrics import recall_score,precision_score,f1_score

Recall = recall_score(y_test,y_predicted_l)
print('Recall:',Recall)

Precision = precision_score(y_test,y_predicted_l)
print('Precision:',Precision)

F1_score = f1_score(y_test,y_predicted_l)
print('F1_score:',F1_score)
```

### Decision Tree Classifier with criterion gini index 

```python
# instantiate the DecisionTreeClassifier model with criterion gini index
clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)


# fit the model
clf_gini.fit(x_train, y_train)
```

```python
y_predict_gini = clf_gini.predict(x_test)
```

#### Check accuracy score with criterion gini index

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score with criterion gini index: {0:0.4f}'. format(accuracy_score(y_test, y_predict_gini)))
```

* Here, y_test are the true class labels and y_pred_gini are the predicted class labels in the test-set.

Compare the train-set and test-set accuracy
Now, I will compare the train-set and test-set accuracy to check for overfitting.

```python
y_pred_train_gini = clf_gini.predict(x_train)

y_pred_train_gini
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_gini)))
```

#### Check for overfitting and underfitting

```python
print('Training set score: {:.4f}'.format(clf_gini.score(x_train, y_train)))

print('Test set score: {:.4f}'.format(clf_gini.score(x_test, y_test)))
```

#### Here, the training-set accuracy score is 0.7708 while the test-set accuracy to be 0.7969. These two values are quite comparable. So, there is no sign of overfitting.
